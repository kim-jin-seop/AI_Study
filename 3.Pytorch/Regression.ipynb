## Linear Regression

```python
import torch
from torch.optim import SGD
```

```python
X_train = torch.tensor([1,2,3])
y_train = torch.tensor([2,4,6])

X_train = X_train.view(-1,1)
y_train = y_train.view(-1,1)

print(X_train, "\n",y_train)
```

```python
W = torch.zeros(1,requires_grad = True)
b = torch.zeros(1,requires_grad = True)
```

```python
optimizer = SGD([W,b], lr=0.01)

nb_epochs = 1000
for epoch in range(1, nb_epochs + 1):
    hypothesis = X_train * W + b
    cost = torch.mean((hypothesis - y_train) ** 2)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    print(epoch, "epoch : ", W, b,cost)
```

## Multi Variable Linear Regression

```python
X_train = torch.tensor([[73,80,75],
                       [93,88,93],
                       [89,91,90],
                       [96,98,100],
                       [73,66,70]], dtype = torch.float)
y_train = torch.tensor([152,185,180,196,142])
y_train = y_train.view(-1,1)
```

```python
W = torch.zeros((3,),requires_grad = True)
b = torch.zeros(1,requires_grad = True)
```

```python
optimizer = SGD([W,b] ,lr = 1e-5)

nb_epochs = 20
for epoch in range(nb_epochs + 1):
    hypothesis = X_train @ W + b
    cost = torch.mean((hypothesis - y_train) ** 2)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch {:4}/{} hypothesis: {}, Cost: {:.6f}'.format(epoch, nb_epochs, hypothesis.squeeze().detach(),cost.item()))
```
