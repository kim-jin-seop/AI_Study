```python
import pandas as pd

df = pd.read_csv('./titanic_train.csv')
df.head()
```

## histogram : 연속값에 대한 구간별 도수 분포 시각화

```python
import matplotlib.pyplot as plt

#나이에 따른 구간별 사람 수
plt.hist(df['Age'])
```

```python
df['Age'].hist()
```

```python
import seaborn as sns
sns.distplot(df['Age'], bins=10, kde = False, rug = True)
```

```python
##
sns.histplot(df['Age'], kde = True)
```

```python
##
plt.figure(figsize=(12,6))
sns.histplot(x='Age', data= df, kde=True, bins=30)
```

## countplot : 카테고리 변수에 대한 건수를 확인 x : 카테고리 변수 y : 그 카테고리의 개수

```python
sns.countplot(df['Pclass'])
```

```python
sns.countplot(x = 'Pclass', data= df)
```

## barplot : 2개의 feature를 확인, x는 이산값, y는 연속값의 평균

```python
sns.barplot(x='Pclass',y='Age',data=df)
```

```python
# y축은 무조건 숫자가 와야함(만약 문자가 오면 수평으로 바뀜)
sns.barplot(x='Pclass',y='Survived',data=df)
```

```python
sns.barplot(x='Pclass',y='Sex',data=df)
```

```python
sns.barplot(x='Pclass',y='Survived',data=df, estimator=sum, ci=80)
```

## bar plot에서 hue 사용

```python
sns.barplot(data=df, x="Pclass",y='Age',hue="Sex")
```

```python
sns.barplot(data= df, x= 'Pclass', y = 'Survived', hue='Sex')
```

```python
sns.barplot(data= df, x= 'Survived', y = 'Pclass', hue='Sex', orient='h')
```

## violin plot

```python
sns.violinplot(y ='Age', data=df)
# 25%, 75% : boxplot이 있누
```

```python
sns.violinplot(x='Pclass',y='Age',data=df)
```

## boxplot : 4분위를 박스 형태로 표시, x가 이산 값, y는 연속값

```python
sns.boxplot(y='Age', data=df)
```

```python
sns.boxplot(x='Pclass', y = 'Age', data=df)
```

## scatter plot : 산포도, x,y가 연속값

```python
sns.scatterplot(x= 'Age', y = 'Fare', data = df)
```

```python
sns.scatterplot(x='Age', y='Fare', data=df, hue='Pclass') #style = 'Survived'
```

# subplot

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(8,4))
sns.violinplot(x='Pclass', y = 'Age', data= df, ax = axes[1])
```

## Heatmap : 주로 연속형 값의 상관관계를 보는 것

```python
df.corr()
```

```python
plt.figure(figsize=(8,8))

corr = df.corr()

sns.heatmap(corr)
```

```python
sns.heatmap(corr, annot=True, fmt= '.1f', cbar=False, linewidths=0.5)
```
